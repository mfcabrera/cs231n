## What are Decision Trees
Decision tree is a ML algorithm that tries to generate a set of rules to predict a value. The value can be a number (regression) or a category (binary or multilabel classification).

Implementations are found in popular ML packages such as scikit-learn where you can basically use them without knowing the internal but before going there I implement an example to understand the inner workings.
